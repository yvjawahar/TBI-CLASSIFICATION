```python
import os
from PIL import Image
import matplotlib.pyplot as plt
```

```python
TBI_folder = '/content/drive/MyDrive/TBI_FINAL_IMAGES/'
NONTBI_folder = '/content/drive/MyDrive/NONTBI_FINAL_IMAGES/'


plt.figure(figsize=(15, 3))
# Process TBI images
for i, filename in enumerate(os.listdir(TBI_folder)):
    if filename.endswith('.png') and i < 5:  # Display only the first 5 images
        img_path = os.path.join(TBI_folder, filename)
        image = Image.open(img_path).convert("RGB")
        plt.subplot(1, 5, i + 1)  # Create a subplot
        plt.imshow(image)
        plt.axis('off')  # Hide axes
plt.figure(figsize=(15, 3))


# Process NONTBI images
for i, filename in enumerate(os.listdir(NONTBI_folder)):
    if filename.endswith('.png') and i < 5:  # Display only the first 5 images
        img_path = os.path.join(NONTBI_folder, filename)
        image = Image.open(img_path).convert("RGB")
        plt.subplot(1, 5, i + 1)  # Create a subplot, starting from index 6
        plt.imshow(image)
        plt.axis('off')  # Hide axes

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
def augment_and_save_images(folder_path, num_augmented_images):
    # Create a directory for augmented images if it doesn't exist
    augmented_folder = os.path.join(folder_path, 'augmented')
    os.makedirs(augmented_folder, exist_ok=True)

    # Process images in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.png'):
            img_path = os.path.join(folder_path, filename)
            image = Image.open(img_path).convert("RGB")

            # Apply augmentation techniques
            for i in range(num_augmented_images):
                augmented_image = image.copy()

                # Random rotation (90 degrees)
                augmented_image_rotated = augmented_image.rotate(90 * i)

                # Random flipping (vertical or horizontal)
                augmented_image_flipped_lr = augmented_image.transpose(Image.FLIP_LEFT_RIGHT)
                augmented_image_flipped_tb = augmented_image.transpose(Image.FLIP_TOP_BOTTOM)

                # Save the augmented images
                augmented_image_rotated.save(os.path.join(augmented_folder, f'{filename.split(".")[0]}_rotated_{i}.png'))
                augmented_image_flipped_lr.save(os.path.join(augmented_folder, f'{filename.split(".")[0]}_flipped_lr_{i}.png'))
                augmented_image_flipped_tb.save(os.path.join(augmented_folder, f'{filename.split(".")[0]}_flipped_tb_{i}.png'))

# Augment and save images in the TBI folder
augment_and_save_images(TBI_folder, num_augmented_images=3)

# Augment and save images in the NONTBI folder
augment_and_save_images(NONTBI_folder, num_augmented_images=3)
```

```python
import os

TBI_folder = '/content/drive/MyDrive/TBI_FINAL_IMAGES/augmented'
NONTBI_folder = '/content/drive/MyDrive/NONTBI_FINAL_IMAGES/augmented'

def count_files_in_folder(folder_path):
    file_count = 0
    for filename in os.listdir(folder_path):
        if os.path.isfile(os.path.join(folder_path, filename)):
            file_count += 1
    return file_count

tbi_file_count = count_files_in_folder(TBI_folder)
nontbi_file_count = count_files_in_folder(NONTBI_folder)

print("Number of files in TBI folder:", tbi_file_count)
print("Number of files in NONTBI folder:", nontbi_file_count)
```
