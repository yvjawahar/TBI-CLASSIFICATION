```python
import os
import pandas as pd
from skimage.feature import greycomatrix, greycoprops
from PIL import Image
import numpy as np
import time
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
def compute_glcm_features(image):
    # Convert image to grayscale
    gray_image = image.convert('L')

    # Compute GLCM
    glcm = greycomatrix(np.array(gray_image), distances=[1], angles=[0], levels=256, symmetric=True, normed=True)

    # Compute GLCM properties
    contrast = greycoprops(glcm, 'contrast')
    correlation = greycoprops(glcm, 'correlation')
    energy = greycoprops(glcm, 'energy')
    homogeneity = greycoprops(glcm, 'homogeneity')

    return contrast[0, 0], correlation[0, 0], energy[0, 0], homogeneity[0, 0]
```

```python
TBI_folder = '/content/drive/MyDrive/TBI_FINAL_IMAGES'
NONTBI_folder = '/content/drive/MyDrive/NONTBI_FINAL_IMAGES'
TBI_folder_augmented = '/content/drive/MyDrive/TBI_FINAL_IMAGES/augmented'
NONTBI_folder_augmented = '/content/drive/MyDrive/NONTBI_FINAL_IMAGES/augmented'

# List to store data
data = []

start_time = time.time()
# Process TBI images
for filename in os.listdir(TBI_folder):
    if filename.endswith('.png'):
        img_path = os.path.join(TBI_folder, filename)
        image = Image.open(img_path).convert("RGB")
        contrast, correlation, energy, homogeneity = compute_glcm_features(image)
        data.append([contrast, correlation, energy, homogeneity,1])
for filename in os.listdir(TBI_folder_augmented):
    if filename.endswith('.png'):
        img_path = os.path.join(TBI_folder_augmented, filename)
        image = Image.open(img_path).convert("RGB")
        contrast, correlation, energy, homogeneity = compute_glcm_features(image)
        data.append([contrast, correlation, energy, homogeneity,1])

# Process NONTBI images
for filename in os.listdir(NONTBI_folder):
    if filename.endswith('.png'):
        img_path = os.path.join(NONTBI_folder, filename)
        image = Image.open(img_path).convert("RGB")
        contrast, correlation, energy, homogeneity = compute_glcm_features(image)
        data.append([contrast, correlation, energy, homogeneity,0])
for filename in os.listdir(NONTBI_folder_augmented):
    if filename.endswith('.png'):
        img_path = os.path.join(NONTBI_folder_augmented, filename)
        image = Image.open(img_path).convert("RGB")
        contrast, correlation, energy, homogeneity = compute_glcm_features(image)
        data.append([contrast, correlation, energy, homogeneity,0])
end_time = time.time()
```

```python
elapsed_time = end_time - start_time
print("Time taken:", elapsed_time, "seconds")
```

```python
print(len(data))
```

```python
df = pd.DataFrame(data, columns=['Contrast', 'Correlation', 'Energy', 'Homogeneity', 'Class'])

# Save DataFrame to CSV
df.to_csv('final_augmented_glcm_features.csv', index=False)
```
